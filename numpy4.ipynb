Bài 1

```python
import numpy as np
a = np.arange(-3, 4, 1, dtype=int)
a1 = np.transpose(a)
b = np.linspace(0, 6, 7)
b1 = np.transpose(b)
c = np.logspace(1.0, 7.0, num=7, base = 2)
c1 = np.transpose(c)
d = np.random.rand(1,7)
d1 = np.transpose(d)
print(a)
print(a.ndim, a1.ndim, a.shape, a1.shape)
print(b)
print(b.ndim, b1.ndim, b.shape, b1.shape)
print(c)
print(c.ndim, c1.ndim, c.shape, c1.shape)
print(d)
print(d.ndim, d1.ndim, d.shape, d1.shape)
```

```python
import numpy as np
a = np.arange(0, 70, 1)
a.shape = (10,7)
a1 = a.flatten()
a2 = a1.reshape(10,7)
b = np.linspace(0, 69, 70)
b.shape = (10,7)
b1 = b.flatten()
b2 = b1.reshape(10,7)
c = np.logspace(0, 10, num=70, base = 2)
c.shape = (10,7)
c1 = c.flatten()
c2 = c1.reshape(10,7)
d = np.random.rand(1, 70)
d.shape = (10,7)
d1 = d.flatten()
d2 = d1.reshape(10,7)
print(a)
print(a1)
print(a2)
print(b)
print(b1)
print(b2)
print(c)
print(c1)
print(c2)
print(d)
print(d1)
print(d2)
```

```python
import numpy as np
def tinh_tb(a): 
    s = 0
    for i in range(0,len(a)): 
        s += a[i]
    tb = s/len(a)
    return tb
a = np.random.randint(0, 50, size = (15,10))
b = np.apply_along_axis(tinh_tb, 0, a)
c = np.apply_along_axis(tinh_tb, 1, a)
print(a)
print("áp dụng trên cột", b)
print("áp dụng trên hàng", c)
```
