```python
import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(-6, 5, num = 1500)
y = x**3 + 3*x + 7
y1 = 3*x**2 + 3
y2 = 6*x

fig, ax = plt.subplots()
ax.plot(x, y, color = "blue", label = "y(x)")
ax.plot(x, y1, color = "red", label = "y'(x)")
ax.plot(x, y2, color = "green", label = 'y"(x)')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
```
